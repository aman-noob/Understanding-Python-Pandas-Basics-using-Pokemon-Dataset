### Sorting

```python
df.sort_values(['attack', 'hp'], ascending = [0,1]).head()
```

### Tweaking Datasets

```python
df['Total'] = df.iloc[:,4:10].sum(axis = 1)
df.head()
```

### Filtering Data

```python
new_df= df.loc[(df['type1'] == 'Grass') & (df['type2'] == 'Poison')].sort_values('attack', ascending = False) # AND
```

```python
new_df = new_df.reset_index(drop = True)
new_df
```

#### Filtering names comtaining Mega

```python
df.loc[~df['name'].str.contains('Mega')]  # (~) drop mega
```

### Regular Expression Lib

```python
import re
df.loc[df['type1'].str.contains('Fire|Grass', flags = re.I, regex=True)]
```

```python
df.loc[df['name'].str.contains('^pi[a-z]*', flags = re.I, regex=True)]  #names starting with Pi usning ^
```

### Conditional changes

```python
df.loc[df['type1']  == 'Fire', 'Legendary'] = True
df.head(3)
```

```python
df.loc[df['Total'] > 500,  ['generation', 'legendary']] == ['T1', 'T2']
df.head()
```

### Aggregate Function

#### Groupby

```python
df.groupby(['type1']).mean().sort_values('hp', ascending = False)
```

#### Count

```python
df['count'] = 1
df.groupby(['type1']).count()['count']
```
